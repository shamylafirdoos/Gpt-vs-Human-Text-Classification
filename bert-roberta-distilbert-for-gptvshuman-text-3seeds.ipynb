```python
# ==========================================================
# Install dependencies
# ==========================================================
!pip install lime shap -q

import pandas as pd
import numpy as np
import torch
import random, time, os, subprocess
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix, f1_score, precision_score,
    recall_score, accuracy_score, brier_score_loss
)
from torch.optim import AdamW
from torch.nn import CrossEntropyLoss
from torch.utils.data import DataLoader, TensorDataset
from statsmodels.stats.contingency_tables import mcnemar
from lime.lime_text import LimeTextExplainer
import shap
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# ==========================================================
# Helper: Seed fixing
# ==========================================================
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

# ==========================================================
# GPU Usage Helper
# ==========================================================
def get_gpu_stats():
    try:
        result = subprocess.check_output(
            ["nvidia-smi", "--query-gpu=name,memory.total,memory.used,utilization.gpu",
             "--format=csv,noheader,nounits"]
        ).decode("utf-8").strip().split("\n")
        stats = []
        for line in result:
            name, total, used, util = line.split(", ")
            stats.append({
                "GPU": name,
                "VRAM_Total_MB": int(total),
                "VRAM_Used_MB": int(used),
                "Utilization_%": int(util)
            })
        return stats
    except Exception:
        return [{"GPU": "CPU", "VRAM_Total_MB": 0, "VRAM_Used_MB": 0, "Utilization_%": 0}]

# ==========================================================
# Tokenization + Train/Eval
# ==========================================================
def tokenize_texts(texts, tokenizer, max_length=256):
    return tokenizer(
        texts.tolist(),
        max_length=max_length,
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    )

def train_epoch(model, loader, optimizer, loss_fn, device): 
    model.train()
    total_loss, total_correct = 0, 0
    for batch in loader:
        input_ids, attention_masks, labels = [x.to(device) for x in batch]
        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_masks, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        preds = torch.argmax(outputs.logits, dim=-1)
        total_correct += (preds == labels).sum().item()
    return total_loss / len(loader), total_correct / len(loader.dataset)

def evaluate_model(model, loader, loss_fn, device): 
    model.eval()
    preds_all, labels_all, probs_all = [], [], []
    total_loss, total_correct = 0, 0
    with torch.no_grad():
        for batch in loader:
            input_ids, attention_masks, labels = [x.to(device) for x in batch]
            outputs = model(input_ids=input_ids, attention_mask=attention_masks, labels=labels)
            total_loss += outputs.loss.item()
            probs = torch.softmax(outputs.logits, dim=-1)[:,1].cpu().numpy()
            preds = torch.argmax(outputs.logits, dim=-1)
            preds_all.extend(preds.cpu().numpy())
            labels_all.extend(labels.cpu().numpy())
            probs_all.extend(probs)
            total_correct += (preds == labels).sum().item()
    return total_loss / len(loader), total_correct / len(loader.dataset), np.array(preds_all), np.array(labels_all), np.array(probs_all)

# ==========================================================
# Calibration Metric (ECE)
# ==========================================================
def expected_calibration_error(probs, labels, n_bins=10):
    bins = np.linspace(0.0, 1.0, n_bins+1)
    ece = 0.0
    for i in range(n_bins):
        mask = (probs >= bins[i]) & (probs < bins[i+1])
        if np.any(mask):
            acc = np.mean(labels[mask] == (probs[mask] >= 0.5))
            conf = np.mean(probs[mask])
            ece += (np.sum(mask) / len(probs)) * abs(acc - conf)
    return ece

# ==========================================================
# Multiple Seeds Experiment
# ==========================================================
def run_experiment(model_name, save_dir, seeds=[7, 42, 123]):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenized = tokenize_texts(df['Data'], tokenizer)
    input_ids = tokenized['input_ids']
    attention_masks = tokenized['attention_mask'] 
    labels = torch.tensor(df['Labels'].values)

    metrics_all = []
    conf_matrices = []
    preds_all, labels_all = [], []
    total_gpu_time = 0

    for seed in seeds:
        set_seed(seed)
        start_time = time.time()

        # Split data
        X_train, X_temp, y_train, y_temp, mask_train, mask_temp = train_test_split(
            input_ids, labels, attention_masks, test_size=0.3, stratify=labels, random_state=seed
        )
        X_val, X_test, y_val, y_test, mask_val, mask_test = train_test_split(
            X_temp, y_temp, mask_temp, test_size=0.5, stratify=y_temp, random_state=seed
        )

        train_data = TensorDataset(X_train, mask_train, y_train)
        val_data = TensorDataset(X_val, mask_val, y_val)
        test_data = TensorDataset(X_test, mask_test, y_test)

        train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
        val_loader = DataLoader(val_data, batch_size=32) 
        test_loader = DataLoader(test_data, batch_size=32)

        # Model
        model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
        model.to(device)
        optimizer = AdamW(model.parameters(), lr=3e-5)
        loss_fn = CrossEntropyLoss()

        # Train
        print(f"\n[{model_name}] Training with Seed={seed}")
        for epoch in range(3):
            train_loss, train_acc = train_epoch(model, train_loader, optimizer, loss_fn, device)
            val_loss, val_acc, _, _, _ = evaluate_model(model, val_loader, loss_fn, device)
            print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f}, Train Acc={train_acc:.4f} | Val Loss={val_loss:.4f}, Val Acc={val_acc:.4f}")

        # Test
        _, test_acc, preds, true_labels, probs = evaluate_model(model, test_loader, loss_fn, device)
        metrics = {
            "Accuracy": accuracy_score(true_labels, preds),
            "Precision": precision_score(true_labels, preds),
            "Recall": recall_score(true_labels, preds),
            "F1": f1_score(true_labels, preds),
            "Brier": brier_score_loss(true_labels, probs),
            "ECE": expected_calibration_error(probs, true_labels)
        }
        metrics_all.append(metrics)
        preds_all.extend(preds)
        labels_all.extend(true_labels)
        conf_matrices.append(confusion_matrix(true_labels, preds))

        # GPU usage
        gpu_stats = get_gpu_stats()
        elapsed = time.time() - start_time
        total_gpu_time += elapsed/3600  # hours
        print(f"[{model_name}] Seed={seed} | GPU Stats: {gpu_stats} | Time (hrs): {elapsed/3600:.4f}")

        # Save fine-tuned model
        safe_name = model_name.replace('/', '-')
        ckpt_dir = os.path.join(save_dir, f"{safe_name}-seed{seed}")
        os.makedirs(ckpt_dir, exist_ok=True)
        model.save_pretrained(ckpt_dir)
        tokenizer.save_pretrained(ckpt_dir)

    return metrics_all, np.array(preds_all), np.array(labels_all), total_gpu_time, tokenizer, model, conf_matrices

# ==========================================================
# CI Calculation (pretty format)
# ==========================================================
def summarize_metrics(metrics_all):
    summary = {}
    for key in metrics_all[0].keys():
        values = [m[key] for m in metrics_all]
        mean = np.mean(values)
        std = np.std(values, ddof=1)
        ci95 = 1.96 * (std / np.sqrt(len(values)))
        summary[key] = f"{mean:.3f} ± {ci95:.3f}"
    return summary

# ==========================================================
# Confusion Matrix with CI
# ==========================================================
def summarize_confusion_matrices(conf_matrices, model_name):
    cms = np.array(conf_matrices)
    mean_cm = np.mean(cms, axis=0)
    std_cm = np.std(cms, axis=0, ddof=1)
    ci95 = 1.96 * (std_cm / np.sqrt(len(cms)))
    
    labels = ["Human", "GPT"]
    plt.figure(figsize=(6,5))
    annot = np.empty_like(mean_cm).astype(str)
    for i in range(mean_cm.shape[0]):
        for j in range(mean_cm.shape[1]):
            annot[i,j] = f"{mean_cm[i,j]:.1f} ± {ci95[i,j]:.1f}"
    sns.heatmap(mean_cm, annot=annot, fmt="", cmap="Blues", xticklabels=labels, yticklabels=labels)
    plt.title(f"Confusion Matrix (mean ± CI) - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

# ==========================================================
# Dataset & Experiments
# ==========================================================
df = pd.read_csv('/kaggle/input/gpt-vs-human-text-dataset/shuffled_data.csv')
df = df[['Data', 'Labels']].dropna()
df = df[df['Data'].astype(str).str.strip().astype(bool)]

save_dir = "./fine_tuned_models"
os.makedirs(save_dir, exist_ok=True)

print("Running BERT...")
bert_metrics, bert_preds, bert_labels, bert_gpu_hours, bert_tokenizer, bert_model, bert_cms = run_experiment("bert-base-uncased", save_dir)

print("Running DistilBERT...")
distilbert_metrics, distilbert_preds, _, distilbert_gpu_hours, distilbert_tokenizer, distilbert_model, distilbert_cms = run_experiment("distilbert-base-uncased", save_dir)

print("Running RoBERTa...")
roberta_metrics, roberta_preds, _, roberta_gpu_hours, roberta_tokenizer, roberta_model, roberta_cms = run_experiment("roberta-base", save_dir)

print("\n=== Results with 95% CIs ===")
print("BERT:", summarize_metrics(bert_metrics))
print("DistilBERT:", summarize_metrics(distilbert_metrics))
print("RoBERTa:", summarize_metrics(roberta_metrics))
print(f"\nGPU Hours Used: BERT={bert_gpu_hours:.3f}, DistilBERT={distilbert_gpu_hours:.3f}, RoBERTa={roberta_gpu_hours:.3f}, Total={bert_gpu_hours+distilbert_gpu_hours+roberta_gpu_hours:.3f}")

# Confusion Matrices with ±
summarize_confusion_matrices(bert_cms, "BERT")
summarize_confusion_matrices(distilbert_cms, "DistilBERT")
summarize_confusion_matrices(roberta_cms, "RoBERTa")

# ==========================================================
# McNemar's Test (pairwise model comparison)
# ==========================================================
print("\n=== McNemar's Test Results ===")

def run_mcnemar_test(labels, preds1, preds2, model1, model2):
    both_correct = np.sum((preds1 == labels) & (preds2 == labels))
    model1_correct_only = np.sum((preds1 == labels) & (preds2 != labels))
    model2_correct_only = np.sum((preds1 != labels) & (preds2 == labels))
    both_wrong = np.sum((preds1 != labels) & (preds2 != labels))

    table = [[both_correct, model1_correct_only],
             [model2_correct_only, both_wrong]]

    result = mcnemar(table, exact=False, correction=True)
    print(f"{model1} vs {model2} -> Statistic={result.statistic:.4f}, p-value={result.pvalue:.4f}")
    if result.pvalue < 0.05:
        print("   ❌ Significant difference at 95% confidence")
    else:
        print("   ✅ No significant difference")

run_mcnemar_test(bert_labels, bert_preds, distilbert_preds, "BERT", "DistilBERT")
run_mcnemar_test(bert_labels, bert_preds, roberta_preds, "BERT", "RoBERTa")
run_mcnemar_test(bert_labels, distilbert_preds, roberta_preds, "DistilBERT", "RoBERTa")
```

```python
# ==========================================================
# LIME visualization (RoBERTa)
# ==========================================================
class TransformersWrapper:
    def __init__(self, model, tokenizer, device):
        self.model = model
        self.tokenizer = tokenizer
        self.device = device

    def predict_proba(self, texts):
        self.model.eval()
        encodings = self.tokenizer(
            texts, truncation=True, padding=True, max_length=256, return_tensors="pt"
        )
        input_ids = encodings["input_ids"].to(self.device)
        attention_mask = encodings["attention_mask"].to(self.device)
        with torch.no_grad():
            outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
            probs = torch.softmax(outputs.logits, dim=-1).cpu().numpy()
        return probs

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
roberta_wrapper = TransformersWrapper(roberta_model, roberta_tokenizer, device)
explainer = LimeTextExplainer(class_names=["Human", "GPT"])

real_texts = [
    "Spyware and adware are types of malicious software designed to collect information about a user's online activities without their knowledge or consent."
]

print("\n=== LIME Explanations (RoBERTa, Visual) ===")
for text in real_texts:
    probs = roberta_wrapper.predict_proba([text])[0]
    pred_class = np.argmax(probs)
    print(f"\nText: {text}\nPredicted Class: {['Human','GPT'][pred_class]} | Probabilities: {probs}")
    exp = explainer.explain_instance(
        text,
        classifier_fn=roberta_wrapper.predict_proba,
        num_features=8
    )
    exp.show_in_notebook(text=True)

# ==========================================================
# SHAP Explainability (RoBERTa, fixed)
# ==========================================================
print("\n=== SHAP Explanations (RoBERTa, Visual) ===")

def shap_predict(texts: list[str]):
    return roberta_wrapper.predict_proba([str(t) for t in texts])

background_texts = df['Data'].dropna().astype(str).sample(min(20, len(df)), random_state=42).tolist()

try:
    masker = shap.maskers.Text()
    shap_explainer = shap.Explainer(shap_predict, masker, output_names=["Human", "GPT"])
    
    real_texts = [str(t) for t in real_texts]
    shap_values = shap_explainer(real_texts)

    for i, text in enumerate(real_texts):
        probs = roberta_wrapper.predict_proba([text])[0]
        pred_class = np.argmax(probs)
        print(f"\n[SHAP] Text: {text}\nPredicted Class: {['Human','GPT'][pred_class]} | Probabilities: {probs}")
        try:
            shap.plots.text(shap_values[i])
        except Exception:
            print("Fallback SHAP values (words, contributions):")
            print(list(zip(shap_values[i].data, shap_values[i].values)))
except Exception as e:
    print("SHAP visualization failed again:", e)
```
